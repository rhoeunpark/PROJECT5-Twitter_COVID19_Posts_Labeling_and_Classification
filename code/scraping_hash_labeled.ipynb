```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Make Request
def post_request(hashtag):
    tweets_ls = []
    
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'covid (#{hashtag}) lang:en').get_items()):
        if i>2000:
            break

        tweets_ls.append([tweet.date, tweet.id, tweet.user.username, tweet.content, tweet.likeCount, tweet.retweetCount])
    
    return tweets_ls
```

```python
# Comsume_posts
def consume_posts(tweets_list):
    columns = ['datetime', 'tweet_id', 'username', 'text', 'likes', 'retweet']
    tweets_df = pd.DataFrame(tweets_list, columns=columns)
    
    return tweets_df
```

```python
# Save data
def save_data(df, overwrite, file):
    if overwrite:
        mode = 'w'
        header = True
    else:
        mode ='a'
        header = False
        
    filepath = file
    
    df.to_csv(filepath,
              mode=mode,
              header=header,
              index=False)
```

```python
# Scraping, Functions put together
def scrape_twitter(hashtags, filepath):
    unique_id = set()
    i = 0
    
    for hashtag in hashtags:
        tweets_ls = post_request(hashtag)
        tweets_df = consume_posts(tweets_ls)
        
        tweets_df = tweets_df[~tweets_df['tweet_id'].isin(unique_id)]
        
        if i == 0:
            overwrite = True
        else: 
            overwrite = False
            
        save_data(tweets_df, overwrite, filepath)
        
        unique_id.update(tweets_df['tweet_id'])
        i += 1
        
        print(hashtag)
        
    print('scraping completed')
```

### Combining two datasets

```python
anti_vax = pd.read_csv("../data/twitter_antivax.csv")
pro_vax = pd.read_csv("../data/twitter_provax.csv")
```

```python
# adding approporiate labels
anti_vax['get_vaccine'] = 0
pro_vax['get_vaccine'] = 1
```

```python
# concatenate dataframe in random order
tweets = pd.concat([anti_vax, pro_vax], axis=0).sample(frac=1, random_state=42).reset_index(drop=True)
```
